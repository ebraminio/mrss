```python
import requests
from urllib.parse import unquote
from tqdm import tqdm
import time
```

```python
for n in tqdm(range(710)):
    r = requests.get(f'https://maktabkhooneh.org/course/a-mk{n}/')
    r.url
    name = unquote(r.url.split('/')[4])
    with open(f'dump/{name}.html', 'wb') as f:
        f.write(r.content)
    time.sleep(1)
```
